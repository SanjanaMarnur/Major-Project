```python
import numpy as np
import pandas as pd
import statistics
```

```python
df = pd.read_csv("wide_dataset.csv")
df
```

```python
df["Seasonal_mean"] = df[["June", "July", "Aug", "Sept", "Oct"]].mean(axis=1)
df
```

```python
df["Seasonal_max"] = df[["June", "July", "Aug", "Sept", "Oct"]].max(axis=1)
df
```

```python
df["Growth_rate"] = df["Sept"] - df["June"]
```

```python
df
```

```python
df["NDVI_range"] = df["Seasonal_max"] - df[["June", "July", "Aug", "Sept", "Oct"]].min(axis=1)
```

```python
df
```

```python
def classify_crop_health(ndvi):
    if ndvi >= 0.33:
        return "Healthy"
    elif ndvi >= 0.18:
        return "Moderate"
    else:
        return "Stressed"

df["Crop_health"] = df["Seasonal_mean"].apply(classify_crop_health)
```
